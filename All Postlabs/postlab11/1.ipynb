```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg")
img_array = np.array(img)

# Padding size (top, bottom, left, right)
pad_top, pad_bottom, pad_left, pad_right = 50, 50, 50, 50

# Add black padding
padded_img = np.pad(
    img_array,
    ((pad_top, pad_bottom), (pad_left, pad_right), (0, 0)),
    mode="constant"
)

# Display original vs padded image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(padded_img)
plt.title("Image with Black Padding")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
img = Image.open(r"D:\Degree Sem 3\PWP\Lab Task\python\Exp 11\MU.jpg")
img_array = np.array(img)

# Separate channels
R = img_array[:, :, 0]
G = img_array[:, :, 1]
B = img_array[:, :, 2]

# Display original and channels
plt.figure(figsize=(12, 6))

plt.subplot(1, 4, 1)
plt.imshow(img_array)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(R, cmap="Reds")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(G, cmap="Greens")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(B, cmap="Blues")
plt.title("Blue Channel")
plt.axis("off")

plt.tight_layout()
plt.show()
```
