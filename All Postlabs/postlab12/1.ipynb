```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
fs = 800
t = np.linspace(0, 1, fs, endpoint=False)
sine1 = np.sin(2 * np.pi * 5 * t)
sine2 = np.sin(2 * np.pi * 12 * t)
added_signal = sine1 + sine2

plt.figure(); plt.plot(t, added_signal, 'k')
plt.title('Post Lab (a): 5 Hz + 12 Hz Sine Waves'); plt.grid(True)
plt.show()
```

```python
fs = 600
t = np.linspace(0, 2, 2*fs, endpoint=False)
sine_wave = np.sin(2 * np.pi * 6 * t)
cosine_wave = np.cos(2 * np.pi * 10 * t)
multiplied_signal = sine_wave * cosine_wave

plt.figure(); plt.plot(t, multiplied_signal, 'c')
plt.title('Post Lab (b): 6 Hz Sine × 10 Hz Cosine'); plt.grid(True)
plt.show()
```

```python
fs = 500
t = np.linspace(0, 1, fs, endpoint=False)
sine_wave = np.sin(2 * np.pi * 8 * t)
shifted_sine = np.sin(2 * np.pi * 8 * (t - 0.15))

plt.figure(); plt.plot(t, sine_wave, label='Original (8 Hz)')
plt.plot(t, shifted_sine, label='Shifted (0.15s)', linestyle='--')
plt.title('Post Lab (c): Time Shift'); plt.legend(); plt.grid(True)
plt.show()
```

```python
# (d) Scale amplitude of 9 Hz sine by factor 2.5
sine9 = np.sin(2 * np.pi * 9 * t)
scaled_sine9 = 2.5 * sine9

plt.figure(); plt.plot(t, sine9, label='Original (9 Hz)')
plt.plot(t, scaled_sine9, label='Scaled ×2.5', linestyle='--')
plt.title('Post Lab (d): Amplitude Scaling'); plt.legend(); plt.grid(True)
plt.show()
```

```python
# (e) Reverse 7 Hz sine in time
sine7 = np.sin(2 * np.pi * 7 * t)
reversed_sine7 = sine7[::-1]

plt.figure(); plt.plot(t, sine7, label='Original (7 Hz)')
plt.plot(t, reversed_sine7, label='Reversed', linestyle='--')
plt.title('Post Lab (e): Time Reversal'); plt.legend(); plt.grid(True)
plt.show()
```
