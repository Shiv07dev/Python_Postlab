```python
# Program: Book Class Example

class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    # method to display book details
    def display(self):
        print(f"Title: {self.title}, Author: {self.author}, Price: {round(self.price, 2)}")

    # method to apply discount
    def apply_discount(self, percent):
        discount_amount = self.price * (percent / 100)
        self.price -= discount_amount


# --- Test Book class ---
print("Enter details for Book 1")
title1 = input("Title: ")
author1 = input("Author: ")
price1 = float(input("Price: "))
book1 = Book(title1, author1, price1)

print("\nEnter details for Book 2")
title2 = input("Title: ")
author2 = input("Author: ")
price2 = float(input("Price: "))
book2 = Book(title2, author2, price2)

print("\nBook Details (before discount):")
book1.display()
book2.display()

# Apply 10% discount to Book 2
book2.apply_discount(10)

print("\nBook Details (after 10% discount on Book 2):")
book1.display()
book2.display()
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def perimeter(self):
        return 2 * math.pi * self.radius


# --- Test Circle class ---
r = float(input("Enter radius of circle: "))
c = Circle(r)
print("Circle radius =", r)
print("Area =", c.area())
print("Perimeter =", c.perimeter())
```
