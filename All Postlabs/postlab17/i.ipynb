```python
import sympy as sp
```

```python
n, z = sp.symbols('n z')
```

```python
x_n = 3**n
X_z = sp.summation(x_n * z**(-n), (n, 0, sp.oo))
print("\nZ-transform of x[n] = 3^n * u[n] is:")
sp.pprint(X_z, use_unicode=True)
```

```python
x_n = (-0.5)**n
X_z = sp.summation(x_n * z**(-n), (n, 0, sp.oo))
print("\nZ-transform of x[n] = (-0.5)^n * u[n] is:")
sp.pprint(X_z, use_unicode=True)
```
