```python
import time

try:
    import serial
    try:
        arduino = serial.Serial(port='COM3', baudrate=115200, timeout=0.1)
        time.sleep(1)
        real_port = True
    except:
        real_port = False
except:
    real_port = False

def write_read(x):
    time.sleep(0.05)
    if real_port:
        arduino.write(bytes(x, 'utf-8'))
        return arduino.readline()
    else:
        try:
            r = int(x) + 1
            return f"b'{r}'"
        except:
            return b''

while True:
    num = input("Enter a Number: ")
    value = write_read(num)
    print(value)
```
