```python
import time

print("POST LAB – Arduino LED Control (Simulation Mode)")
print("------------------------------------------------\n")
print("Type ON or OFF to control the LED. Type EXIT to quit.\n")

while True:
    command = input("Enter command (ON/OFF): ").strip().upper()

    if command == "EXIT":
        print("\nSimulation ended. Connection closed.")
        break

    if command == "ON":
        time.sleep(0.3)
        print("Sent: ON")
        print("Arduino Response: LED Turned ON\n")

    elif command == "OFF":
        time.sleep(0.3)
        print("Sent: OFF")
        print("Arduino Response: LED Turned OFF\n")

    else:
        print("Invalid command! Please type ON or OFF.\n")
    


```
