```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Mohe', 'Karnal', 'Yrik', 'Jack'],
    'Age': [30, 21, 29, 28]
}

df = pd.DataFrame(data)
```

```python
sns.boxplot(df['Age'])
plt.title("Age Distribution - Box Plot")
plt.show()
```

```python
sns.violinplot(df['Age'])
plt.title("Age Distribution - Violin Plot")
plt.show()
```
